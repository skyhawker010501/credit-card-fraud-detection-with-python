```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load a dataset towards pandas dataframe
cc_data = pd.read_csv('/content/creditcard.csv')
```

```python
#print first five rows
cc_data.head()
```

```python
cc_data.tail()
```

```python
cc_data
```

```python
#dataset information
cc_data.info()
```

```python
# disribution of legit and fraud transaction
cc_data['Class'].value_counts()
```

```python
#this dataset is highly unbalanced
#0--> Normal Transaction and 1-->Fraudlent Transaction

#seprating the data for analysis
legal = cc_data[cc_data.Class==0]
fraud = cc_data[cc_data.Class==1]
```

```python
print(legal.shape)
print(fraud.shape)
```

```python
#statical measure of data
legal.Amount.describe()
#fraud.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the value of both transaction
cc_data.groupby('Class').mean()
```

```python
#Under Sampling
#-->means a build sample dataset contaning similar distrubution of legal and fraud transation
legal_sample = legal.sample(n=156)
```

```python
new_dataset = pd.concat([legal_sample,fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x = new_dataset.drop(columns = 'Class', axis=1)
y = new_dataset['Class']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, stratify = y, random_state =2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training


```python
#logistic regression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#accuracy prediction
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print(training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(testing_data_accuracy)
```
